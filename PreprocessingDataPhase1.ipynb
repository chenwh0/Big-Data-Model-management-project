```python
import regex
import pandas
```

## **1. Load data**


```python
FILE_PATH = "/content/drive/MyDrive/2026 Spring/Principles of Big Data & Model Management/Project/tweets.json"
```

```python
raw_data = pandas.read_json(FILE_PATH, lines=True) # file contains many JSON objects, each new JSON object is on newline, so line=True
raw_data
```

## **2. Extract only needed columns**

```python
interested_data = raw_data[["id", "entities"]]
interested_data
```

## **3. Extract only hashtags & URLs**

```python
hashtags = []
urls = []
for row in interested_data.itertuples():
    hashtags.append(row.entities["hashtags"])
    urls.append(row.entities["urls"])
```

```python
final_data = interested_data.copy()
final_data["urls"] = urls
final_data["hashtags"] = hashtags
final_data
```

```python
final_data.drop("entities", axis=1, inplace=True) # drop entities column
final_data
```

## **4. Store preprocessed data**

```python
final_data.to_csv("/content/drive/MyDrive/2026 Spring/Principles of Big Data & Model Management/Project/hashtags_and_urls_only_tweets.csv")
```
