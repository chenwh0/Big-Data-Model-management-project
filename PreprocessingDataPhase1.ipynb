```python
import pandas
```

## **1. Load data**


```python
FILE_PATH = r"C:\Users\Jack\Documents\School\cs8540 Project\Big-Data-Model-management-project\tweets.json"
```

```python
raw_data = pandas.read_json(FILE_PATH, lines=True) # file contains many JSON objects, each new JSON object is on newline, so line=True
raw_data
```

## **2. Extract only needed columns**

```python
interested_data = raw_data[["entities"]]
interested_data
```

## **3. Extract only hashtags & URLs**

```python
hashtags = []
urls = []
for tweet in interested_data.itertuples():
    tweet_hashtags = []
    tweet_urls = []
    if tweet.entities["hashtags"] != []:
        for hashtag in tweet.entities["hashtags"]:
            tweet_hashtags.append(hashtag["text"])

    if tweet.entities["urls"] != []:
        for url in tweet.entities["urls"]:
            tweet_urls.append(url["url"])

    hashtags.append(tweet_hashtags)
    urls.append(tweet_urls)
```

```python
final_data = interested_data.copy()
final_data["urls"] = urls
final_data["hashtags"] = hashtags
final_data
```

```python
final_data.drop("entities", axis=1, inplace=True) # drop entities column
final_data
```

```python
import ast

#need to make everything into a list
def make_list(x):
    if(isinstance(x, str)):
        return ast.literal_eval(x)
    elif isinstance(x, list):
        return x
    else:
        return []
    
    
final_data["urls"] = final_data["urls"].apply(make_list)
final_data["hashtags"] = final_data["hashtags"].apply(make_list)

#remove empty lists
final_data = final_data[final_data["urls"].str.len() > 0]
final_data = final_data[final_data["hashtags"].str.len() > 0]

#need to join list with spaces
final_data["urls"] = final_data["urls"].apply(lambda x: ' '.join(x))
final_data["hashtags"] = final_data["hashtags"].apply(lambda x: ' '.join(x))
```

## **4. Store preprocessed data**

```python
final_data.to_csv(r"C:\Users\Jack\Documents\School\cs8540 Project\Big-Data-Model-management-project\csv_version.csv", index=False) # Drop index columns
```
